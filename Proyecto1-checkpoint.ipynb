# Limpieza de Datos

Se aplicaran las siguiente transformaciones al data set:

- **Eliminar valores nulos** en columnas clave.  
- **Convertir `duration_ms` a minutos** para una interpretación más sencilla.  
- **Asegurar `track_album_release_date` en formato fecha** y extraer el año de lanzamiento.  
- **Normalizar artistas** separando colaboraciones en filas individuales. 


```python
df_clean = df_raw.copy()
```

## Eliminar valores nulos
Se eliminó el único valor nulo que habíamos encontrado en esta fila (gracias a df_raw.isnull().sum()) para asegurar una mayor consistencia en los datos

```python
df_clean = df_clean.dropna(subset=["track_album_name"])
```

## Convertir duración de milisegundos a minutos

```python
df_clean["duration_ms"] = df_clean["duration_ms"]/60000
df_clean = df_clean.rename(columns={"duration_ms":"duration_min"})
```

## Asegurar track_album_release_date en formato fecha

```python
df_clean["track_album_release_date"] = pd.to_datetime(df_clean["track_album_release_date"],errors="coerce")

df_clean["year"] = df_clean["track_album_release_date"].dt.year ##Agregamos también un campoque contiene el año de lanzamiento
```

## Verificamos transformaciones

```python
df_clean.info()
```

```python
df_clean[["track_album_release_date", "year", "duration_min"]].head()
```

```python
df_clean.describe()
```

## Normalizar artistas
Se creó un nuevo dataframe para observar la frecuencia con la que aparece cada artista, incluyendo cuando aparecen en una colaboración

```python
df_long = df_clean.copy()

df_long["artist"] = df_long["track_artist"].str.split(",")
df_long = df_long.explode("artist")
df_long["artist"] = df_long["artist"].str.strip()
```

```python
df_long.head()
```

<h1>Visualizaciones y Hallazgos</h1>

```python
df_long["artist"].value_counts().head(10).plot(kind="barh")
plt.title("Top 10 artistas con más canciones (incluyendo colaboraciones)")
plt.xlabel("Número de canciones")
plt.show()
```

```python
df_clean["duration_min"].plot(kind="hist", bins=30, edgecolor="black")
plt.title("Distribución de la duración de canciones")
plt.xlabel("Duración (minutos)")
plt.show()
```

```python
df_clean.plot.scatter(x="energy", y="danceability", alpha=0.5)
plt.title("Energía vs Bailabilidad")
plt.show()
```

```python
df_clean.groupby("year")["track_popularity"].mean().plot(kind="line", marker="o")
plt.title("Popularidad promedio de canciones por año")
plt.ylabel("Popularidad promedio")
plt.xlabel("Año")
plt.show()
```

<h1>Conclusiones</h1>

**Distribución temporal**: El dataset abarca canciones desde 1954 hasta la actualidad, aunque la mayoría proviene de las últimas décadas.

**Duración de canciones**: La mayoría de canciones populares duran entre 3 y 4 minutos, lo que coincide con el estándar de la industria musical.

**Artistas más presentes**: Algunos artistas (ej. Bad Bunny, Drake, Dua Lipa) aparecen repetidamente, tanto en solitario como en colaboraciones, confirmando su fuerte presencia en las listas de popularidad.

**Energía y bailabilidad**: Existe una tendencia a que las canciones populares se concentren en valores medios-altos de energía y bailabilidad, lo que sugiere que ambos atributos influyen en el atractivo comercial.
